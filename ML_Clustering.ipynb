## K Means and K Medoids

```python
df = pd.read_csv('Seed_Data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df)
```

```python
from sklearn.cluster import KMeans

k = [3,5,7]
for i in k:
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df)
    df['k_means_'+str(i)] = kmeans.labels_
```

```python
df
```

```python
!pip install scikit-learn-extra
```

```python
from sklearn_extra.cluster import KMedoids
```

```python
for j in k:
    kmedoids = KMedoids(n_clusters=j,random_state=0)
    kmedoids.fit(df)
    df['kmedoids_cluster_'+str(j)] = kmedoids.labels_
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['A'],df['P'],c = df['k_means_7'],cmap='viridis')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.show()
```

```python
from sklearn import metrics
```

```python
for i in k:
    print("K-Mean Clustering")
    print("K Value = ",i)

    print("Silhouette Score: ", metrics.silhouette_score(df,df['k_means_'+str(i)]))
    print("Rand Score: ", metrics.rand_score(df['target'],df['k_means_'+str(i)]))
    print("Adjusted Rand Score: ", metrics.adjusted_rand_score(df['target'],df['k_means_'+str(i)]))
    print("Davies Bouldin Score: ", metrics.davies_bouldin_score(df,df['k_means_'+str(i)]))
```

```python
for i in k:
    print("K-Medoid Clustering")
    print("K Value = ",i)

    print("Silhouette Score: ", metrics.silhouette_score(df,df['kmedoids_cluster_'+str(i)]))
    print("Rand Score: ", metrics.rand_score(df['target'],df['kmedoids_cluster_'+str(i)]))
    print("Adjusted Rand Score: ", metrics.adjusted_rand_score(df['target'],df['kmedoids_cluster_'+str(i)]))
    print("Davies Bouldin Score: ", metrics.davies_bouldin_score(df,df['kmedoids_cluster_'+str(i)]))
```

## Hierarchial Clustering and DBSCAN

```python
from sklearn.cluster import AgglomerativeClustering,DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
X = df.iloc[:,[3,4]].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df.describe()
```

```python
h = AgglomerativeClustering(n_clusters=5)
h_clusters = h.fit_predict(X_scaled)
h_clusters
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
print("Hierarchial Clustering")
print("Silhouette Score: ",metrics.silhouette_score(X,h_clusters))
print("Davies Bouldin Score: ",metrics.davies_bouldin_score(X,h_clusters))
```

```python
print("DBSCAN Clustering")
print("Silhouette Score: ",metrics.silhouette_score(X,dbscan_labels))
print("Davies Bouldin Score: ",metrics.davies_bouldin_score(X,dbscan_labels))
```

```python
plt.figure(figsize=(6,8))
sns.scatterplot(x=X[:,0],y=X[:,1],hue=h_clusters,palette="viridis")
plt.title("Hierarchial Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
plt.figure(figsize=(6,8))
sns.scatterplot(x=X[:,0],y=X[:,-1],hue=dbscan_labels,palette="viridis")
plt.title("Hierarchial Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
